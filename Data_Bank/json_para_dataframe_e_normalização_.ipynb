## 3- Normalização dos itens dessa coluna de dicionário e sua divisão em dois dataframes separados, seguindo o modelo relacional.

Normalização é o processo de modelar o banco de dados projetando a forma como as informações serão armazenadas a fim de eliminar, ou pelo menos minimizar, redundância e anomalias no banco de dados.
A terceira forma normal é considerada o nível mínimo necessário para grande parte das aplicações, por essa razão descrevi definições de normalização até esse nível.

* Primeira Forma Normal (1FN): todos os seus atributos são simples e
monovalorados.

* Segunda Forma Normal (2FN):Evita dependência funcional (parcial), ou seja, os registros na tabela, que não são chaves, dependem da chave primária em sua totalidade e não apenas parte dela.
* Terceira Forma Normal (3FN):Evita dependência transitiva, ou seja, ao analisarmos uma tupla (linha) e não encontramos nenhum atributo não chave dependente de outro atributo não chave.

Observação: para uma relação atender as exigências de uma forma normal, se faz necessário que esta obedeça as regras da forma normal anterior.

```python
# Tabela de notas fiscais

df_NFE= df[['CreateDate', 'EmissionDate', 'Discount', 'NFeNumber', 'NFeID']]
df_NFE
```

Esta tabela está na 3FN e tem NFeID ou NFeNumber como chave primária.

```python
# Tabela com o conteúdo das notas fiscais

NF_content = pd.json_normalize(data, record_path =['ItemList'],meta=[['NFeID']])
NF_content
```

Considerando a hipótese de que, a depender da nota fiscal, o preço do produto pode mudar, e tendo como chave primária a combinação de NFeID e ProdutName (não acontece no exemplo, mas um mesmo produto poderia estar em outra nota fiscal), pode-se dizer que a tabela está normalizada.
Porém, nesta tabela podem ocorrer problemas de atualização. Por exemplo, se tivéssemos que atualizar o nome do produto, teriámos que fazê-lo em todas as linhas.
Vamos então criar um Id para o produto e manter Id e nome do produto em uma tabela separada.

```python
# Criação da Tabela de produtos

Product = NF_content['ProductName'].unique() # extrai, em formato de array, os nome únicos de produtos da tabela NF_content 
Product = pd.DataFrame(Product, columns = ['ProductName'])
Product.insert(0, 'ProductID', range(1, len(Product)+1)) # Inserir coluna ProductID contendo números incrementais
Product
```

Esta tabela está na 3FN e tem ProductID como chave primária.

```python
# Nova tabela com o conteúdo das notas fiscais substituindo ProductName for ProductID

NF_content_new = NF_content.merge(Product,on='ProductName') # Mesclar as tabelas  NF_content e Product através da coluna 'ProductName'
NF_content_new  = NF_content_new.drop('ProductName', axis=1) # Excluir coluna 'ProductName'
NF_content_new = NF_content_new.iloc[:, ::-1] # Reverter ordem das colunas do dataframe
NF_content_new
```

Esta tabela está na 3FN. Possui uma chave primária composta ( ProductID e NFeIF) e possui 2 chaves estrangeiras: ProductID  a relacionando com a tabela de produtos, e NFeID a relacionando com a tabela de notas fiscais.
